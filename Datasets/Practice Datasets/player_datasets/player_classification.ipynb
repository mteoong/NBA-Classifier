```python
import pandas as pd
pd.options.mode.chained_assignment = None
```

```python
t2019 = pd.read_csv('./player_datasets/2019-2020.csv')
t2018 = pd.read_csv('./player_datasets/2018-2019.csv')
a2018 = pd.read_csv('./player_datasets/2018-2019-A.csv').drop(["W", "L", "MIN"], axis=1)
p2018 = pd.read_csv('./player_datasets/2018-2019-P36.csv').drop(["Team", "Age", "W", "L", "MIN"], axis=1)


p2018
```

```python
t2018
```

```python
def getFantasyRankIncrease(currYear, nextYear):
    result = pd.merge(currYear, nextYear, how="inner", on=["Player"])
    fantasy = result[["Player","#_x", "#_y"]]
    fantasyRankIncrease = fantasy["#_x"] - fantasy["#_y"]
    fantasy['FPR_INCREASE'] = fantasyRankIncrease
    fantasy = fantasy.drop(['#_x', '#_y'], axis=1)
    fantasy = fantasy.sort_values('FPR_INCREASE', ascending=False)
    result2 = pd.merge(p2018, fantasy, how="inner", on=["Player"])
    return result2
p2018 = getFantasyRankIncrease(t2018, t2019)
p2018
```

```python
for stringg in p2018.head():
    #t2018.plot.scatter(x=stringg, y="FPR_INCREASE")
    if (stringg != "Player" and stringg != "Team"):
        a = p2018[stringg]
        b = p2018["FPR_INCREASE"]
        correlation = a.corr(b)
        print(stringg + " : " + str(correlation))
```
